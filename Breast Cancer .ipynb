# Decision Trees

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
clean_data = data.copy()
clean_data['diagnosis_label'] = (clean_data['diagnosis'])
y=clean_data[['diagnosis_label']].copy()
y.head()
```

```python
cats = ['radius_mean', 'texture_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 
        'compactness_worst', 'concavity_worst']

X = clean_data[cats].copy()
X.columns
y.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=324)
X_train.head()
y_train.describe()
```

```python
diagnosis_classifier = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)
diagnosis_classifier.fit(X_train, y_train)
```

```python
predictions = diagnosis_classifier.predict(X_test)

predictions[:10]

y_test['diagnosis_label'][:10]
```

```python
accuracy_score(y_true = y_test, y_pred = predictions)
```
